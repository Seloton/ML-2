```python
from sklearn.datasets import load_iris
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
iris = load_iris()
```

```python
# нормирование входных данных: (x - min) / (max - min)

X = (iris.data - np.min( iris.data, axis = 0 )) / ( np.max( iris.data, axis = 0 ) - np.min( iris.data, axis = 0 ) )
```

```python
ohe = OneHotEncoder( sparse_output = False )
Y = ohe.fit_transform( iris.target.reshape(-1,1) )

# выполнить кодирование one-hot можно также с помощью pandas:
# Y = pd.get_dummies( iris.target ).to_numpy()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, train_size = 0.75 )
```

```python
from keras.models import Model, Sequential
from keras.layers import Input, Dense
import matplotlib.pyplot as plt
```

```python
x = Input( shape = (4,) )

l1 = Dense( units = 8, activation = "relu" )( x )

out = Dense( units = 3, activation = "sigmoid" )( l1 )

nn = Model(x, out)

"""
Эквивалентная запись:
nn = Sequential( [
      Dense( units = 8, activation = "relu", input_shape = (4,) ),
      Dense( units = 3, activation = "sigmoid" )
])


Для задачи регрессии меняется выходной слой. Обычно один линейный нейрон
nn = Sequential( [
      Dense( units = 8, activation = "relu", input_shape = (4,) ),
      Dense( units = 1, activation = "linear" ) #в приниципе, activation = "linear" можно не указывать - эта активационная функция используется по умолчанию
])
"""
```

```python
nn.compile( optimizer = "adam", loss = "categorical_crossentropy", metrics = ["categorical_accuracy"] )

# можно попробовать такие методы обучения как "rmsprop" (используется по умолчанию), "SGD" и посмотреть на эффективность обучения
# !для задачи регрессии нужна другая функция потерь!!! например, средний квадрат ошибки: loss = "mse" или средняя абсолютная "mae". метрика будет та же - можно ее не указывать
```

```python
hist = nn.fit( X_train, Y_train, epochs = 100, batch_size = 4 )

# ! если хотим обучить модель снова (а не дообучить), то ее необходимо снова создать и пересобрать
```

```python
plt.plot( hist.history["loss"], label = "Loss" )
plt.plot( hist.history["categorical_accuracy"], label = "Accuracy" )
plt.xlabel("Epochs")
plt.ylabel("Loss / Accuracy")
plt.legend()
plt.show()
```

```python
eval = nn.evaluate( X_test, Y_test )
```

```python
y_pred = nn.predict( X_test )
y_pred
```

```python
yp = np.argmax( y_pred, axis = 1 )
yp
```

```python
yt = np.argmax( Y_test, axis = 1 )
yt
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score( yt, yp )
```
